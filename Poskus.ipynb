```python
from sage.graphs.graph_generators import graphs

# Določimo število vozlišč
n = 3

# Pretvorimo generator v seznam, da lahko dostopamo do vseh grafov
all_graphs = list(graphs(n))

# Izrišemo vsak graf posebej
for i, G in enumerate(all_graphs):
    print(f"Graf {i + 1}:")
    G.show()
```
